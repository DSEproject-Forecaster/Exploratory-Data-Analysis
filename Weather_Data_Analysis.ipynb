#Neural Prophet Temperature

```python
!pip install neuralprophet
from neuralprophet import NeuralProphet
```

```python
df = pd.read_csv("/content/drive/MyDrive/Sem5/DSE Project/DataSet/Weather.csv")[['time_stamp', 'temperature']]

df.time_stamp = pd.to_datetime(df.time_stamp)
# df = df.loc[(df.time_stamp >= '2018-01-01') & (df.time_stamp <= '2021-05-30')]
df.rename(columns = {'time_stamp':'ds', 'temperature':'y'}, inplace = True)
```

```python
m = NeuralProphet(changepoints_range=0.95, n_changepoints=50, trend_reg=1, weekly_seasonality=False, daily_seasonality=10, yearly_seasonality=10)
```

```python
df['ds'] = pd.DatetimeIndex(df['ds'])
df.head()
```

```python
df_train, df_val = m.split_df(df, freq='H', valid_p = 0.2)
metrics = m.fit(df_train, freq='H', validation_df=df_val)
```

```python
seasonal_components = m.predict_seasonal_components(df)
seasonal_components
```

```python
seasonal_components.daily[:25]
```

```python
metrics
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(metrics["MAE"], '-o', label="Training Loss")  
ax.plot(metrics["MAE_val"], '-r', label="Validation Loss")
ax.legend(loc='center right', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set_xlabel("Epoch", fontsize=28)
ax.set_ylabel("Loss", fontsize=28)
ax.set_title("Model Loss (MAE)", fontsize=28)
```

```python
future = m.make_future_dataframe(df, periods=24*365*2, n_historic_predictions=len(df))
future.tail()
```

```python
forecast = m.predict(future)
```

#Neural Prophet Dewpoint

```python
# !pip install neuralprophet
from neuralprophet import NeuralProphet
```

```python
df = pd.read_csv("/content/drive/MyDrive/Sem5/DSE Project/DataSet/Weather.csv")[['time_stamp', 'dewpoint_temperature']]

df.time_stamp = pd.to_datetime(df.time_stamp)
# df = df.loc[(df.time_stamp >= '2018-01-01') & (df.time_stamp <= '2021-05-30')]
df.rename(columns = {'time_stamp':'ds', 'dewpoint_temperature':'y'}, inplace = True)
```

```python
df['ds'] = pd.DatetimeIndex(df['ds'])
df.head()
```

```python
m = NeuralProphet(changepoints_range=0.95, n_changepoints=50, trend_reg=1, weekly_seasonality=False, daily_seasonality=10)
df_train, df_val = m.split_df(df, freq='H', valid_p = 0.2)
metrics = m.fit(df_train, freq='H', validation_df=df_val)
```

```python
metrics
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.plot(metrics["MAE"], '-o', label="Training Loss")  
ax.plot(metrics["MAE_val"], '-r', label="Validation Loss")
ax.legend(loc='center right', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.set_xlabel("Epoch", fontsize=28)
ax.set_ylabel("Loss", fontsize=28)
ax.set_title("Model Loss (MAE)", fontsize=28)
```

```python
future = m.make_future_dataframe(df, periods=24*365*2, n_historic_predictions=len(df))
forecast = m.predict(future)
```

```python
forecast
```

```python
m.plot_components(forecast)
plt.show()
```

```python
forecast_plot = m.plot(forecast)
plt.show()
```
